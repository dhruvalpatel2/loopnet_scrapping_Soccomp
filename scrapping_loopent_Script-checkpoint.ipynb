```python
import csv
import random
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException

# Function to introduce random delays
def random_delay():
    time.sleep(random.uniform(2, 5))  # Random delay between 2 and 5 seconds

# Initialize the WebDriver
driver = webdriver.Chrome()  # Make sure to set the correct path to your chromedriver
wait = WebDriverWait(driver, 10)

# Navigate to the initial page and scrape the URLs
driver.get('https://www.loopnet.com/search/commercial-real-estate/usa/for-sale/?sk=fd44917fd63887f2dea28566916665ab')
random_delay()  # Random delay before loading URLs
wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'article.placard a')))

# Use a set to store URLs to avoid duplicates
urls = set()
for link in driver.find_elements(By.CSS_SELECTOR, 'article.placard a'):
    url = link.get_attribute('href')
    if url.startswith('https://www.loopnet.com/'):
        urls.add(url)

# Dictionary to hold all property details
all_property_details = []

# Iterate over the set of URLs
for url in urls:
    try:
        driver.get(url)
        random_delay()  # Random delay before looking for elements
        # Wait for the property facts table to load
        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'table.property-data')))
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        # Initialize a dictionary to hold the details for this property
        property_details = {'Property URL': url}

        # Use BeautifulSoup to extract information
        facts_table = soup.find('table', class_='property-data')
        if facts_table:
            for row in facts_table.find_all('tr'):
                cells = row.find_all('td')
                if len(cells) == 2:  # Ensure there is a key-value pair
                    fact_type = cells[0].get_text(strip=True)
                    fact_value = cells[1].get_text(strip=True)
                    property_details[fact_type] = fact_value

        # Append this property's details to the all_property_details list
        all_property_details.append(property_details)
        print(f"Scraped property details from {url}")
    except TimeoutException:
        print(f"Timeout occurred while trying to access {url}")
        continue  # Skip to the next URL

# Close the driver when done
driver.quit()

# Writing to CSV
csv_file = 'properties.csv'
# Find all unique fact types to create the CSV column headers
fact_types = set(ft for details in all_property_details for ft in details)

with open(csv_file, 'w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=['Property URL'] + sorted(fact_types))
    writer.writeheader()
    for property_details in all_property_details:
        writer.writerow(property_details)

print(f"Data has been written to {csv_file}")
```
